```python
import pandas as pd
import pickle as p
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

import seaborn as sns
```

```python
df = pd.read_csv('data/diet_data.csv')
df.head()
```

```python
sns.heatmap(df.isnull())
```

```python
df = df.drop(columns='Date')
```

```python
df.dropna(inplace=True)
```

```python
sns.heatmap(df.isna(), cbar=False)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1:]
```

```python
x.head()
```

```python
st = StandardScaler()
```

```python
x = st.fit_transform(x)
```

```python
#y = st.fit_transform(y)
```

```python
xtrain , xtest, ytrain, ytest = train_test_split(x, y, test_size=0.1 ,random_state=123)
```

```python
xtrain.shape, xtest.shape
```

```python
ytrain.shape, ytest.shape
```

```python
lr = LinearRegression(normalize=True).fit(xtrain, ytrain)
```

```python
ypred = lr.predict(xtest)
```

```python
lr.score(xtest, ytest)
```

```python
p.dump(st, open('standardScaler.pk', 'wb'))
p.dump(lr, open('model.pk', 'wb'))
```
